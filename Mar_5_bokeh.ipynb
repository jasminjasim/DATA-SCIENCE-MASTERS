# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
# glyphs is the base class for creating different shapes.
# to call in bokeh,first we need to create a figure object.using this object
# we can add different plots using glyphs
# eg:
import seaborn as sns
tips=sns.load_dataset('tips')
from bokeh.plotting import figure,show
fig=figure(title='new pic')
fig.circle(x=tips['tip'],y=tips['total_bill'])
show(fig)
```

# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips
```

```python
# eg:


from bokeh.plotting import figure,show
fig=figure(title='new pic')
fig.xaxis.axis_label='tip'
fig.yaxis.axis_label='total_bill'

fig.circle(x=tips['tip'],y=tips['total_bill'],legend_label='objects',fill_color='red')
fig.line(x=tips['total_bill'],y=tips['size'],legend_label='size')
show(fig)
```
